```python
from google_play_scraper import app, reviews
import pandas as pd
import os
```

#### Scraping ulasana aplikasi Brimo

scraping di Play Store yakni aplikasi Brimo

[Brimo](https://play.google.com/store/apps/details?id=id.co.bri.brimo)

```python
# Menentukan ID aplikasi Brimo
app_id = 'id.co.bri.brimo'

# Mengambil ulasan
reviews_data, _ = reviews(
    app_id,
    lang='id',
    country='id',
    count=10000
)

# Konversi ke DataFrame
reviews_df = pd.DataFrame(reviews_data)
reviews_df
```

```python
# save to csv
reviews_df.to_csv('brimo_reviews.csv', index=False)
```
